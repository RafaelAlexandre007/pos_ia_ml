```python
import pandas as pd

dataset = pd.read_csv('Pos MLs04b Celsius para Fahrenheit.csv')
dataset
```

```python
import seaborn as sns

sns.scatterplot(x=dataset['Celsius'], y=dataset['Fahrenheit'])
```

```python
X_train = dataset['Celsius']
y_train = dataset['Fahrenheit']
```

```python
X_train.shape
```

```python
import tensorflow as tf

model = tf.keras.Sequential()
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=1, input_shape=[1]))
model.add(tf.keras.layers.Dense(units=1))
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.1), loss="mean_absolute_error")
epochs_hist = model.fit(X_train, y_train, epochs=500)
```

```python
import matplotlib.pyplot as plt

plt.plot(epochs_hist.history['loss'])
```

```python
model.get_weights()
```

```python
9/5
```

```python
y_pred = model.predict(X_train)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_train, y_pred)
```

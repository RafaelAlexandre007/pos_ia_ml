```python
import pandas as pd
dataset = pd.read_csv('Pos MLs04d bike share.csv')
dataset
```

```python
dataset = dataset.drop(columns=['casual', 'registered', 'instant'], axis=1)
dataset
```

```python
dataset.dteday = pd.to_datetime(dataset.dteday, format='%Y-%m-%d')
dataset
```

```python
dataset.index = pd.DatetimeIndex(dataset.dteday)
dataset
```

```python
dataset = dataset.drop(columns=['dteday'], axis=1)
dataset
```

```python
dataset['cnt'].asfreq('W').plot()
```

```python
dataset_categorical = dataset[['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit']]

for column in dataset_categorical:
    print(f"{column} unique values {dataset_categorical[column].nunique()}")
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
dataset_categorical = encoder.fit_transform(dataset_categorical).toarray()
dataset_categorical
```

```python
dataset_categorical.shape
```

```python
dataset_categorical = pd.DataFrame(dataset_categorical)
dataset_categorical
```

```python
dataset_numerical.shape
```

```python
dataset_numerical = dataset_numerical.reset_index()
```

```python
dataset_numerical
```

```python
X_dataset = pd.concat([dataset_categorical, dataset_numerical], axis=1)
X_dataset
```

```python
X_dataset = X_dataset.drop(columns="dteday", axis=1)
X_dataset
```
